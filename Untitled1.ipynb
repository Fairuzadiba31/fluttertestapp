```python
import matplotlib.pyplot as plt

A = (1, 1)
B = (0, 0)
C = (1, 0)

shx, shy = 2, 2

A_sx = (A[0] + shx * A[1], A[1])
B_sx = (B[0] + shx * B[1], B[1])
C_sx = (C[0] + shx * C[1], C[1])

A_sy = (A[0], A[1] + shy * A[0])
B_sy = (B[0], B[1] + shy * B[0])
C_sy = (C[0], C[1] + shy * C[0])

#plotting
original_points = [A, B, C, A]
sheared_x_points = [A_sx, B_sx, C_sx, A_sx]
sheared_y_points = [A_sy, B_sy, C_sy, A_sy]

#Separate coordinates
def get_xy(points):
    return [p[0] for p in points], [p[1] for p in points]

orig_x, orig_y = get_xy(original_points)
sx_x, sx_y = get_xy(sheared_x_points)
sy_x, sy_y = get_xy(sheared_y_points)

plt.figure(figsize=(6,6))

plt.plot(orig_x, orig_y, marker='o', color='purple', linewidth=1.5, markersize=6, label='Original Triangle')
plt.plot(sx_x, sx_y, marker='o', color='red', linewidth=1.5, markersize=6, label='Sheared X-axis')
plt.plot(sy_x, sy_y, marker='o', color='green', linewidth=1.5, markersize=6, label='Sheared Y-axis')

# Labels
for i, (x, y) in enumerate(original_points[:-1]):
    plt.text(x + 0.05, y + 0.05, f"P{i}({x},{y})", fontsize=12, color='purple')

for i, (x, y) in enumerate(sheared_x_points[:-1]):
    plt.text(x - 0.05, y - 0.15, f"P{i}'({x},{y})", fontsize=12, color='red')

for i, (x, y) in enumerate(sheared_y_points[:-1]):
    plt.text(x + 0.05, y - 0.1, f"P{i}''({x},{y})", fontsize=12, color='green')

# Axes and grid
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.grid(True, linestyle='--', alpha=0.5)
plt.axis('equal')
plt.legend()
plt.title("Shearing along X-axis and Y-axis (Single Figure)")
plt.show()

print("Coordinates after shearing on X-axis:")
print(f"A' = {A_sx}, B' = {B_sx}, C' = {C_sx}")
print("Coordinates after shearing on Y-axis:")
print(f"A'' = {A_sy}, B'' = {B_sy}, C'' = {C_sy}")
```

```python
import matplotlib.pyplot as plt

A = (2, 3)
B = (4, 2)
C = (4, 5)

A_x = (A[0], -A[1])
B_x = (B[0], -B[1])
C_x = (C[0], -C[1])

A_y = (-A[0], A[1])
B_y = (-B[0], B[1])
C_y = (-C[0], C[1])

#plotting
original_points = [A, B, C, A]
reflected_x_points = [A_x, B_x, C_x, A_x]
reflected_y_points = [A_y, B_y, C_y, A_y]

# Separate coordinates
def get_xy(points):
    return [p[0] for p in points], [p[1] for p in points]

orig_x, orig_y = get_xy(original_points)
rx_x, rx_y = get_xy(reflected_x_points)
ry_x, ry_y = get_xy(reflected_y_points)

plt.figure(figsize=(6,6))

plt.plot(orig_x, orig_y, marker='o', color='purple', linewidth=1.5, markersize=6, label='Original Triangle')
plt.plot(rx_x, rx_y, marker='o', color='red', linewidth=1.5, markersize=6, label='Reflected on X-axis')
plt.plot(ry_x, ry_y, marker='o', color='green', linewidth=1.5, markersize=6, label='Reflected on Y-axis')

# Labels
for i, (x, y) in enumerate(original_points[:-1]):
    plt.text(x + 0.2, y - 0.1, f"P{i}({x},{y})", fontsize=12, color='purple')

for i, (x, y) in enumerate(reflected_x_points[:-1]):
    plt.text(x + 0.2, y - 0.25, f"P{i}'({x},{y})", fontsize=12, color='red')

for i, (x, y) in enumerate(reflected_y_points[:-1]):
    plt.text(x + 0.3, y - 0.1, f"P{i}''({x},{y})", fontsize=12, color='green')

# Axes and grid
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.grid(True, linestyle='--', alpha=0.5)
plt.axis('equal')
plt.legend()
plt.title("Reflection of Triangle on X-axis and Y-axis (Single Figure)")
plt.show()
```
